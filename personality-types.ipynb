```python
import pandas as pd
import numpy as np
import math
```

```python
##Introversion
print("Introversion")
tn_i =  brigg[(brigg.IE == 'E') &  (brigg.IEAn == 'E')].shape[0]
tp_i =  brigg[(brigg.IE == 'I') &  (brigg.IEAn == 'I')].shape[0]
fn_i = brigg[(brigg.IE == 'I') &  (brigg.IEAn == 'E')].shape[0]
fp_i = brigg[(brigg.IE == 'E') &  (brigg.IEAn == 'I')].shape[0]

p_i = tp_i + fn_i
n_i = fp_i + tn_i
fpr_i = fp_i / n_i

print(tn_i)
print(tp_i)
print(fn_i)
print(fp_i)


prev_i = p_i / (p_i + n_i)
acc_i = (tp_i + tn_i) / (p_i + n_i) 
sen_i = tp_i / p_i
spe_i = tn_i / n_i
mcc_i = ((tp_i * tn_i) - (fp_i * fn_i)) / (math.sqrt((tp_i + fp_i)*(p_i)*(n_i)*(tn_i+fn_i)))
prev_thres_i = math.sqrt(fpr_i) / (math.sqrt(sen_i) + math.sqrt(fpr_i)) 
jacc_i = tp_i / (tp_i + fn_i + fp_i)
prec_i = tp_i / (tp_i + fp_i)
fow_i = math.sqrt((prec_i * sen_i))
f1_i = 2 * (prec_i * sen_i) / (prec_i + sen_i)
ba_i = (sen_i + spe_i) / 2
info_i = (sen_i + spe_i) - 1


print("Prevalence: ", prev_i)
print("Accuracy: ", acc_i)
print('Sensitivity: ', sen_i)
print('Specificity: ', spe_i)
print("MCC: ", mcc_i)
print("Prevalence Threshold: ", prev_thres_i)
print('Jaccard Index: ', jacc_i)
print('Precision: ', prec_i)
print('Fowlkes-Mallow: ', fow_i) 
print('F1 Score: ', f1_i)
print('Balanced Accurarcy: ', ba_i)
print('Informedness: ', info_i)
```
